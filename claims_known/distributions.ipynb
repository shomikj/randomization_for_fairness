```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sympy.stats import QuadraticU, density
import math
```

```python
plt.rcParams["figure.figsize"] = [4,3]
fig, ax = plt.subplots()

x = np.linspace(0,1,1000)
y = np.ones(1000)
plt.plot(x,y,color="#1f77b4")

plt.ylim([0,2])

plt.ylabel("Density")
plt.xlabel("Claims ($\,c_i\,$)")

#plt.legend()
#plt.grid()
plt.tight_layout()
plt.savefig('uniform_dist.eps', format='eps', dpi=1000)
plt.show()
```

```python
plt.rcParams["figure.figsize"] = [4,3]
fig, ax = plt.subplots()

x = np.linspace(0,1,1000)

loc, scale = 0.5, 0.1
a, b = (0 - loc) / scale, (1 - loc) / scale
y1 = stats.truncnorm(a, b, loc=loc, scale=scale).pdf(x)
#y1 = np.max(y1)-y1

loc, scale = 0.5, 0.15
a, b = (0 - loc) / scale, (1 - loc) / scale
y2 = stats.truncnorm(a, b, loc=loc, scale=scale).pdf(x)
#y2 = np.max(y2)-y2

loc, scale = 0.5, 0.2
a, b = (0 - loc) / scale, (1 - loc) / scale
y3 = stats.truncnorm(a, b, loc=loc, scale=scale).pdf(x)
#y3 = np.max(y3)-y3

plt.plot(x,y1,color="#bc5090", label="0.10")
plt.plot(x,y2,color="#1f77b4", label="0.15")
plt.plot(x,y3,color="#ffa600", label="0.20")

#plt.ylim([0,4])

plt.ylabel("Density")
plt.xlabel("Claims ($\\,c_i\\,$)")

plt.legend(title="Distribution $\\sigma$")
#plt.grid()
plt.tight_layout()
plt.savefig('normal_dist.jpg', format='jpg', dpi=500)
plt.show()
```
