```python
df = df[df["treatment6"]=="no program"]
```

```python
features = [
    "age",
    "canton_moth_tongue",
    "city_big",
    "city_medium",
    "city_no",
    "emp_share_last_2yrs",
    "emp_spells_5yrs",
    "female",
    "foreigner_b",
    "foreigner_c",
    "gdp_pc",
    "married",
    "other_mother_tongue",
    "past_income",
    "prev_job_manager",
    "prev_job_sec_mis",
    "prev_job_sec1",
    "prev_job_sec2",
    "prev_job_sec3",
    "prev_job_self",
    "prev_job_skilled",
    "prev_job_unskilled",
    "qual_semiskilled",
    "qual_degree",
    "qual_unskilled",
    "qual_wo_degree",
    "swiss",
    "ue_spells_last_2yrs",
    "unemp_rate",
    "employed12",
]

df = df[features]
```

```python
df.head()
```

```python
df["unemp_rate"]
```

```python
# Min Max Normalization for Continuous Features

continuous_features = ["age", "emp_spells_5yrs", "past_income", "ue_spells_last_2yrs", "unemp_rate"]

for f in continuous_features:
    df[f] = (df[f]-df[f].min()) / (df[f].max() - df[f].min())
```

```python
df.to_csv("../../../data/swiss_data/swiss_data_clean.csv", index=False)
```
